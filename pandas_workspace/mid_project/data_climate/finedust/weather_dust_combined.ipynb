```python
dust_seoul = total_df[total_df['시도코드'] == 11]
dust_seoul
```

```python
def month(x):
    return re.compile('(\d{4})(\d{2})(\d{2})').sub('\g<1>-\g<2>', str(x))

dust_seoul['YYYY-MM']=dust_seoul['일시'].apply(month)
dust_seoul['일시']=dust_seoul['일시'].astype('str')
dust_seoul
```

```python
f, ax = plt.subplots(figsize = (20, 7))
sns.lineplot(x = '일시',
             y = 'PM10',
              data = dust_seoul)

plt.xlabel('YYYY-MM', fontsize=12 )
plt.ylabel('미세먼지농도(PM10)', fontsize=12)
plt.title('2016 전국 월별 미세먼지농도', fontsize=16)
```
