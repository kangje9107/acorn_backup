```python
def sido_code_func(vec1):
    for i in sido_code:
        if vec1 == sido_code[i]:
            return i
            pass
        pass
    pass


def dust_grade_func(m):
    if m <= 30 : 
        return 1
    elif m <=80 :
        return 2
    elif m < 150 :
        return 3
    elif m >= 150:
        return 4
    else:
        return m.index.values()
        pass
    pass
pass

def fdust_grade_func(n):
    if n <= 15 : 
        return 1
    elif n <= 50 :
        return 2
    elif n < 100 :
        return 3
    elif n >=100:
        return 4
    else:
        return n.index.values()
        pass
    pass
pass

```
