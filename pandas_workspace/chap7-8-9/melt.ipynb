```python
import pandas as pd
```

```python
# pd.melt() method
#                                                          <converted>
# 1) The list of specified columns's names of a dataframe -------------> 'variable'(new column created automatically)'s data
#                                                          <converted>
# 2) The specified column's data of a dataframe           -------------> 'value'(new column created automatically)'s data
```

```python
df_weather = pd.read_csv('../doit_pandas/data/weather.csv')
```

```python
df_weather.head()
```

```python
df_weather.info()
```

```python
help(pd.melt)
```

```python
df_weather_melt = pd.melt(df_weather, id_vars=['element'])
df_weather_melt
# 고정시킨 element 열을 제외한, 나머지 모든 열의 이름을 'variable' 열의 데이터로 변환하고
# 고정시킨 element 열을 제외한, 나머지 모든 열의 데이터를 'value' 열의 데이터로 변환
```

```python
# df_weather_melt[df_weather_melt['variable'] == 'id'].count()
# df_weather_melt[df_weather_melt['variable'] == 'd1'].count()
df_weather_melt[df_weather_melt['variable'] == 'd31'].count()
```

```python
df_weather_melt = pd.melt(df_weather, id_vars='element', var_name='day', value_name='temp')
# 자동으로 추가된 variable, value 열의 이름을 명시적으로 지정

df_weather_melt
```

```python
df_weather_melt = pd.melt(df_weather, id_vars=['id', 'year', 'month'], var_name='day', value_name='temp')
# id_vars 인자에 지정한 열들을 제외한, 나머지 열들의 이름은 -> variable 열의 데이터로 변환
# id_Vars 인자에 지정한 열들을 제외한, 나머지 열들의 값은 -> value 열의 데이터로 변환

df_weather_melt
```

```python
df_weather_melt[ df_weather_melt['day'] == 'element' ].count()
```

```python
df_weather_melt[ df_weather_melt['day'] == 'element' ]
```

```python
df_weather_melt = pd.melt(df_weather, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temperature')
# id_vars 인자에 지정한 열들을 제외한, 나머지 열들의 이름은 --> 'variable' 열의 데이터로 변환
# id_vars 인자에 지정한 열들을 제외한, 나머지 열들의 값은 --> 'value' 열의 데이터로 변환

df_weather_melt
```
