```python
# naver api 로 블로그 검색 결과 가져오기
import os
import sys
import urllib.request
import json

temp_list = []
client_id = "네이버 API ID"
client_secret = "네이버 API 코드"
encText = urllib.parse.quote("스프레이건 설치 후기") #검색어 

start_num = 1  #검색 블로그 포스트 각 넘버링 / 최대 1000번까지 가능
total_list = [] #총 검색결과 담을 list


for num in range(11):
    if (start_num== 1000):
        start_num = 1001   #start_num 최대 1000까지 가능하므로
    url = "https://openapi.naver.com/v1/search/blog?&start="+str(start_num)+"&display=100&query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        json_data = json.loads(response_body) #네이버 Open API를 통해서 수신된 데이터가 JSON 포멧이기 때문에, JSON 형태 데이터를  dict 형태로 만듦
        total_list.append(json_data["items"])  #list 안에 list(각 블로그 포스트 100개씩)
        start_num += 100
    else:
        print("Error Code:" + rescode)
```

```python
# api에서 가져온 결과 txt 파일로 정리(링크 등)
output = open("스프레이건_결과3.txt", 'w', encoding='utf-8')
num=1
for total in total_list:
    for post in total:
        old_title = post["title"]
        new_title = old_title.replace("<b>","").replace("</b>","")
        old_link = post["link"]
        if "naver" not in old_link:  #검색 결과중에 네이버 블로그 아닌것들 거르기
            continue
        new_link = old_link.replace("?Redirect=Log&amp;logNo=","/") 
        old_descrip = post["description"]
        new_descrip = old_descrip.replace("<b>","").replace("</b>","").replace("...","")
        print(new_link)
        print(new_title)
        print("------------------------------------------")
        output.write('{}\t'.format(new_link))
        output.write('{}\t'.format(post["postdate"]))
        output.write('{}\t'.format(new_title))
        output.write('{}\t'.format(new_descrip))
        output.write('\n')
        print(num)
        num+=1
output.close()
```

```python
# 링크 for문 돌려서 크롤링

from selenium import webdriver  
from bs4 import BeautifulSoup as bs
import requests 
driver = webdriver.Chrome('C:/chrome/chromedriver.exe')
driver.implicitly_wait(3)           # 암묵적으로 웹 자원 로드를 위해 3초까지 기다려 준다.


f = open('스프레이건_결과3.txt', 'r', encoding='utf-8') #link 담겨 있는 파일 로드
result = open("스프레이건_결과4.txt", 'w', encoding='utf-8') # 결과 파일 만들기
lines = f.readlines() # line by line 읽어줌 
num = 1
tem_list=[]

for line in lines:
    new_line = ""
    print(num)
    link = line.split("\t")[0].replace('"','').replace('\ufeff','')     
    date = line.split("\t")[1]
    driver.get(link) #셀레니움  드라이버로 url 접근 
    driver.switch_to_frame("mainFrame")  #naver 블로그 html 안에 iframe 형태로 되어 있어서 다시 들어가주어야 원하는 html 로드 가능
    html = driver.page_source
    soup = bs(html, 'html.parser')
    
    ### 각 블로그마다 html태그가 조금씩 다르게 생김. 일단 발견한 요소는 다음과 같은데 더 있을 수도..ㅠㅠㅠㅠㅠ
    if (soup.select('.se-module-text')!=[]):   
        text = soup.select('.se-module-text')
    elif (soup.select('#postViewArea')!=[]):
        text = soup.select('#postViewArea')
    elif (soup.select('.se_textarea > span')!=[]):
        text = soup.select('.se_textarea > span')
    else:
        text = soup.select('.se_textarea')
    # text 라인별로 추출된거 다 합쳐주기 + 불필요한 문자 제거
    for i in text:
        new_line = new_line + i.text.replace("\u200b","").replace("\n","").replace("\xa0","") 
    new_line.splitlines()
    tem_list.append(new_line)
    result.write('{}\t'.format(num))
    result.write('{}\t'.format(link))
    result.write('{}\t'.format(new_line))
    result.write('{}\t'.format(date))

    result.write('\n')        
    #print(new_line)  전체 게시글 텍스트
    print("================================")
    num += 1
result.close()
```
